### Pre-procesamiento

- Usaremos una técnica para estandarizar datos numéricos usando opciones de escala y centrado.
- Los valores cercanos a la media se transformarán en valores cercanos a 0.
- Los valores cercanos a la desviación estándar recibirán valores cercanos a 1.

```python
from sklearn.preprocessing import StandardScaler

# fit(X[, y]) - Compute the mean and std to be used for later scaling.
scaler = StandardScaler().fit(X)
print(scaler)
```

```python
# transform(X[, y, copy]) - Perform standardization by centering and scaling
X_transf = scaler.transform(X)
print(X_transf)
```

```python
df.head()
```

```python
df_transf = pd.DataFrame(X_transf, columns=cols[0:8])
df_transf.head()
```

### Guardando los datos

```python
import numpy as np

np.save('X.npy', X)
np.save('y.npy', y)
np.save('X_transf.npy', X_transf)
```
