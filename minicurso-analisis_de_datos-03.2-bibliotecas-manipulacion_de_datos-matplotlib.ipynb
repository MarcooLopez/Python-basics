### 3.2. Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
y1 = np.random.rand(50)
plt.plot(y1, marker='o');

# x = range(50)
# plt.plot(x, y1, marker='o');
```

```python
# Misma grafica
x = range(50)
plt.plot(x, y1, marker='o');
```

```python
plt.plot(x, y1, marker='o', color="orange")
plt.grid()
```

```python
x=np.random.normal(4,1,50)
y1=10+x+np.random.normal(0,1,50)
plt.scatter(x, y1, marker='.');
```

```python
y2=12+x+np.random.normal(0,1,50)
y3=15+x+np.random.normal(0,1,50)
plt.scatter(x, y1, marker='.')
plt.scatter(x, y2, marker='.')
plt.scatter(x, y3, marker='.')
plt.xlabel("Tiempo (s)")
plt.ylabel("Temperatura (C)")
plt.legend(["Round 1", "Round 2", "Round 3"])
plt.show();
```

```python
x=["Domingo","Lunes","Martes","Miercoles","Jueves","Viernes"]
y=np.random.normal(10,4,6)
plt.bar(x, y);
```

#### Generando gráfico en PDF

```python
%config InlineBackend.figure_format = 'pdf'
plt.bar(x, y);
```

#### Gerando gráfico em formato PNG

```python
%config InlineBackend.figure_format = 'png'
plt.bar(x, y);
```

```python
anhos = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
pib = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
plt.plot(anhos, pib, marker='o')
plt.title('PIB')
plt.xlabel('Año')
plt.ylabel('Billones de pesos$')
plt.grid()
```
